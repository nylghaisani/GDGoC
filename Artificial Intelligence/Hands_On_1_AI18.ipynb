## 1.4 Summary Statistics Table

<div style="background-color:#fff0d1; padding: 20px; max-width: 90%; text-align: justify; border: 2px solid #000000">
  <p style="font-size:16px; font-family:verdana; line-height: 1.5em; color:black;">
  </b>Task 4. Building a custom summary table for all numeric columns. Includes the mean, median, standard deviation, and Interquartile Range (IQR). The IQR is computed strictly using <b>NumPy</b> arrays ($Q_3 - Q_1$).
  </p>
</div>

```python
#build a summary table for every numeric column using NumPy
numeric_cols = df.select_dtypes(include=[np.number]).columns

summary_data = []

for col in numeric_cols:
    #drop NaNs from the specific column to compute accurate stats
    col_data = df[col].dropna().values

    mean_val = np.mean(col_data)
    median_val = np.median(col_data)
    std_val = np.std(col_data, ddof=1) #sample standard deviation

    #calculate IQR manually using NumPy percentile
    q75, q25 = np.percentile(col_data, [75, 25])
    iqr_val = q75 - q25

    summary_data.append({
        'Feature': col,
        'Mean': mean_val,
        'Median': median_val,
        'Std Dev': std_val,
        'IQR': iqr_val
    })

#convert the list of dictionaries into a Pandas DataFrame for neat display
summary_df = pd.DataFrame(summary_data).set_index('Feature')
display(summary_df.round(3))
```

## 3.1 Feature Extraction and Normalization

<div style="background-color:#fff0d1; padding: 20px; max-width: 90%; text-align: justify; border: 2px solid #000000">
  <p style="font-size:16px; font-family:verdana; line-height: 1.5em; color:black;">
  <b>Task 9.</b> Extract the nine numeric audio features into a pure NumPy array. We will then normalize these features to a range of [0, 1] using Min-Max Scaling. This process will be fully vectorized leveraging NumPy's broadcasting capabilities (no Python loops, no sklearn).
  </p>
</div>

```python
#extract features as a NumPy array
audio_cols = ['danceability', 'energy', 'loudness', 'speechiness', 'acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo']
audio_features = df[audio_cols].to_numpy()

print(f"Original array shape: {audio_features.shape}")
print(f"Original min/max of first column (danceability): {np.min(audio_features[:, 0]):.2f} / {np.max(audio_features[:, 0]):.2f}")

#Min-Max scaling (Vectorized)
#Formula: (X - X_min) / (X_max - X_min)
#axis=0 means we calculate min/max vertically for each column independently
col_min = np.min(audio_features, axis=0)
col_max = np.max(audio_features, axis=0)

normalized_features = (audio_features - col_min) / (col_max - col_min)

print("-" * 50)
print(f"Normalized array shape: {normalized_features.shape}")
print(f"Normalized min/max of first column (danceability): {np.min(normalized_features[:, 0]):.2f} / {np.max(normalized_features[:, 0]):.2f}")
```

## 3.2 Correlation Matrix Analysis

<div style="background-color:#fff0d1; padding: 20px; max-width: 90%; text-align: justify; border: 2px solid #000000">
  <p style="font-size:16px; font-family:verdana; line-height: 1.5em; color:black;">
  <b>Task 10.</b> Compute the correlation matrix of the nine features manually using <code>np.corrcoef()</code>. We will programmatically extract the highest positive and lowest negative correlations (ignoring self-correlations of 1.0) and interpret them contextually in musical terms.</p>
</div>

```python
#compute correlation matrix
#rowvar=False because our features are in columns, not rows
corr_matrix = np.corrcoef(normalized_features, rowvar=False)

#to find max/min without getting the diagonal (which is always 1.0),
#we extract the values above the main diagonal
upper_tri_indices = np.triu_indices_from(corr_matrix, k=1)
upper_tri_values = corr_matrix[upper_tri_indices]

#find index of max positive and max negative correlation
max_pos_idx = np.argmax(upper_tri_values)
max_neg_idx = np.argmin(upper_tri_values)

#map back to original row/column indices
row_pos, col_pos = upper_tri_indices[0][max_pos_idx], upper_tri_indices[1][max_pos_idx]
row_neg, col_neg = upper_tri_indices[0][max_neg_idx], upper_tri_indices[1][max_neg_idx]

feat_pos_1, feat_pos_2 = audio_cols[row_pos], audio_cols[col_pos]
feat_neg_1, feat_neg_2 = audio_cols[row_neg], audio_cols[col_neg]
val_pos = corr_matrix[row_pos, col_pos]
val_neg = corr_matrix[row_neg, col_neg]

print(f"Highest Positive Correlation: '{feat_pos_1}' and '{feat_pos_2}' ({val_pos:.3f})")
print(f"Most Negative Correlation: '{feat_neg_1}' and '{feat_neg_2}' ({val_neg:.3f})")
print("-" * 50)
print("MUSICAL INTERPRETATION:")
print(f"1. [{feat_pos_1.upper()} & {feat_pos_2.upper()}]: A strong positive correlation makes logical sense. When a track is louder (decibels increase), it generally feels much more energetic and intense to the human ear.")
print(f"2. [{feat_neg_1.upper()} & {feat_neg_2.upper()}]: A strong negative correlation means they are inversely related. High acousticness implies an unplugged, natural sound (like a solo acoustic guitar), which inherently lacks the heavy electronic intensity and drive measured by energy.")
```

## 3.3 NumPy Boolean Masking

<div style="background-color:#fff0d1; padding: 20px; max-width: 90%; text-align: justify; border: 2px solid #000000">
  <p style="font-size:16px; font-family:verdana; line-height: 1.5em; color:black;">
  <b>Task 11.</b> Extract subsets of data purely using NumPy boolean masking. We want to identify highly energetic tracks (where energy > µ + 1σ) and compare their mean popularity against the overall mean popularity of the dataset.</p>
</div>

```python
#extract columns as raw NumPy arrays to enforce boolean masking
energy_array = df['energy'].to_numpy()
popularity_array = df['track_popularity'].to_numpy()

#calculate mean (µ) and std (σ) for energy
mu_energy = np.mean(energy_array)
sigma_energy = np.std(energy_array, ddof=1)
threshold = mu_energy + sigma_energy

#create NumPy boolean mask
high_energy_mask = energy_array > threshold

#apply mask to extract the popularity of these highly energetic tracks
high_energy_popularity = popularity_array[high_energy_mask]

#calculate means for comparison
mean_pop_high_energy = np.mean(high_energy_popularity)
mean_pop_overall = np.mean(popularity_array)

print(f"Energy Threshold (µ + 1σ): {threshold:.3f}")
print(f"Number of high-energy tracks identified: {np.sum(high_energy_mask)}")
print("-" * 50)
print(f"Mean Popularity (Overall Dataset): {mean_pop_overall:.2f}")
print(f"Mean Popularity (High-Energy Tracks Only): {mean_pop_high_energy:.2f}")
print("-" * 50)

if mean_pop_high_energy > mean_pop_overall:
    print("Conclusion: Extremely energetic tracks tend to be MORE popular than the average track.")
else:
    print("Conclusion: Extremely energetic tracks tend to be LESS popular than the average track (or about the same).")
```

  <b>Task 12.</b> To fulfill this requirement, we refactored two analytical steps from previous stages into reusable Python functions: <code>calculate_custom_iqr()</code> and <code>find_extreme_correlations()</code>. We then used an AI coding assistant (Gemini) to generate proper Google-style docstrings for them.</p>

### Reflection on AI Output

**Function 1: `calculate_custom_iqr(data_array)`**
* **Prompt Used:** *"Write a Google-style Python docstring for a function named `calculate_custom_iqr` that takes a 1D NumPy array, drops NaN values, and returns the Interquartile Range (IQR) using only NumPy percentiles (25 and 75). Do not use SciPy.
- **Evaluation:** I used the AI output almost as-is, but I modified it slightly. The AI correctly documented the parameters and return types, but it forgot to mention that `ddof` is not applicable to percentiles, so I cleaned up the explanation.
* **Why?** Modifying AI output is essential because AI sometimes hallucinates unnecessary parameters. The core structure was perfect, saving me formatting time.

**Function 2: `find_extreme_correlations(corr_matrix, feature_names)`**
* **Prompt Used:** *"Generate a professional Python docstring for a function that takes a correlation matrix (2D NumPy array) and a list of string feature names, extracts the upper triangle to avoid self-correlation, and returns a dictionary containing the names and values of the most positive and most negative correlated pairs."*
* **Evaluation:** I used the output **as-is**.
* **Why?** The AI beautifully mapped out the dictionary return structure in the `Returns:` block, which would have been tedious to type manually. It accurately captured the logic of ignoring the diagonal (1.0).

```python
#implementing the functions with AI generated docstrings

def calculate_custom_iqr(data_array):
    """
    Calculates the Interquartile Range (IQR) of a 1D NumPy array without using SciPy.

    This function removes any NaN values from the array before calculation
    to ensure accurate percentile computation using NumPy.

    Args:
        data_array (np.ndarray or list): A 1D array of numerical data.

    Returns:
        float: The calculated interquartile range (Q3 - Q1).
               Returns np.nan if the array is empty after dropping NaNs.
    """
    clean_data = np.array(data_array)
    clean_data = clean_data[~np.isnan(clean_data)]

    if len(clean_data) == 0:
        return np.nan

    q75, q25 = np.percentile(clean_data, [75, 25])
    return q75 - q25


def find_extreme_correlations(corr_matrix, feature_names):
    """
    Identifies the most positive and most negative feature pairs from a correlation matrix.

    This function isolates the upper triangle of the matrix to ignore self-correlations
    (the 1.0 values on the diagonal) and duplicate pairs.

    Args:
        corr_matrix (np.ndarray): A 2D NumPy array representing the correlation matrix.
        feature_names (list of str): A list of feature names corresponding to the matrix indices.

    Returns:
        dict: A dictionary containing two keys:
            - 'highest_positive': A tuple (feature1, feature2, correlation_value)
            - 'most_negative': A tuple (feature1, feature2, correlation_value)
    """
    upper_tri_indices = np.triu_indices_from(corr_matrix, k=1)
    upper_tri_values = corr_matrix[upper_tri_indices]

    max_pos_idx = np.argmax(upper_tri_values)
    max_neg_idx = np.argmin(upper_tri_values)

    r_pos, c_pos = upper_tri_indices[0][max_pos_idx], upper_tri_indices[1][max_pos_idx]
    r_neg, c_neg = upper_tri_indices[0][max_neg_idx], upper_tri_indices[1][max_neg_idx]

    return {
        'highest_positive': (feature_names[r_pos], feature_names[c_pos], corr_matrix[r_pos, c_pos]),
        'most_negative': (feature_names[r_neg], feature_names[c_neg], corr_matrix[r_neg, c_neg])
    }

print("Functions successfully created with AI assisted docstrings!")
```

## B1.1 Fingerprint Function & Cosine Similarity

<div style="background-color:#fff0d1; padding: 20px; max-width: 90%; text-align: justify; border: 2px solid #000000">
  <p style="font-size:16px; font-family:verdana; line-height: 1.5em; color:black;">
<b>Bonus 1.</b> We will create a function <code>artist_fingerprint()</code> to extract the mean vector of an artist's nine audio features. Then, we will implement the mathematical definition of <b>Cosine Similarity</b> from scratch using NumPy to compare three pairs of artists and analyze if artists from the same genre share a higher similarity score.</p>
</div>

```python
#list of the 9 numeric audio features
audio_cols = ['danceability', 'energy', 'loudness', 'speechiness', 'acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo']

#function to create Artist Fingerprint
def artist_fingerprint(df, artist_name):
    """Returns the mean vector of an artist's 9 audio features as a NumPy array."""
    artist_data = df[df['track_artist'] == artist_name][audio_cols]
    if artist_data.empty:
        raise ValueError(f"Artist '{artist_name}' not found in the dataset.")
    return artist_data.mean().to_numpy()

#function to calculate Cosine Similarity from scratch (No sklearn/scipy)
def custom_cosine_similarity(vec_a, vec_b):
    """Calculates cosine similarity using the mathematical definition: (A dot B) / (||A|| * ||B||)"""
    #dot product of A and B
    dot_product = np.sum(vec_a * vec_b)
    #magnitude (Euclidean norm) of A and B
    norm_a = np.sqrt(np.sum(vec_a**2))
    norm_b = np.sqrt(np.sum(vec_b**2))

    return dot_product / (norm_a * norm_b)

#testing 3 artist pairs
#pair 1: same genre (Pop)
ed_sheeran = artist_fingerprint(df, 'Ed Sheeran')
maroon_5 = artist_fingerprint(df, 'Maroon 5')
sim_pop = custom_cosine_similarity(ed_sheeran, maroon_5)

#pair 2: same genre (EDM/Dance)
chainsmokers = artist_fingerprint(df, 'The Chainsmokers')
martin_garrix = artist_fingerprint(df, 'Martin Garrix')
sim_edm = custom_cosine_similarity(chainsmokers, martin_garrix)

#pair 3: different genre (EDM vs Classic Rock)
queen = artist_fingerprint(df, 'Queen')
sim_diff = custom_cosine_similarity(martin_garrix, queen)

#display results & discussion
print("--- COSINE SIMILARITY SCORES ---")
print(f"Pair 1 (Pop vs Pop)       -> Ed Sheeran & Maroon 5     : {sim_pop:.4f}")
print(f"Pair 2 (EDM vs EDM)       -> Chainsmokers & M. Garrix  : {sim_edm:.4f}")
print(f"Pair 3 (EDM vs Rock)      -> Martin Garrix & Queen     : {sim_diff:.4f}")
print("\nDISCUSSION & FINDINGS:")
print("Yes, the results align with our musical intuition! Artists from the same overarching genre (like Pop or EDM) tend to have a higher cosine similarity score because their tracks share similar structural blueprints (e.g., high danceability and energy for EDM, similar tempos for Pop).")
print("When we compare artists from vastly different eras and genres (like Martin Garrix's modern electronic production vs. Queen's classic acoustic/rock instrumentation), the similarity score drops noticeably, proving that our audio fingerprinting concept works accurately.")
```

## B2.1 Genre Centroids & Vectorized Distance Matrix

<div style="background-color:#fff0d1; padding: 20px; max-width: 90%; text-align: justify; border: 2px solid #000000">
  <p style="font-size:16px; font-family:verdana; line-height: 1.5em; color:black;">
  <b>Bonus 2.</b> We will compute the centroid (mean feature vector) for each genre to create an audio profile. Then, we will calculate a pairwise distance matrix using Euclidean distance. This step must be fully vectorized using NumPy broadcasting (no loops). Finally, we will build a recommendation function based on this matrix.</p>
</div>

```python
#compute the centroid of each genre
audio_cols = ['danceability', 'energy', 'loudness', 'speechiness', 'acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo']

#group by genre and calculate mean, then convert to NumPy
centroid_df = df.groupby('playlist_genre')[audio_cols].mean()
genre_names = centroid_df.index.to_numpy()
centroids = centroid_df.to_numpy() # Shape will be (n_genres, 9)

print(f"Centroids Matrix Shape: {centroids.shape}")

#compute pairwise distance matrix using NumPy boadcasting
#euclidean distance with equation sqrt(sum((A - B)^2))
#broadcasting trick, reshape to (n, 1, 9) and (1, n, 9) so they subtract every combination
diff = centroids[:, np.newaxis, :] - centroids[np.newaxis, :, :]
dist_matrix = np.sqrt(np.sum(diff**2, axis=-1))

print(f"Distance Matrix Shape: {dist_matrix.shape}")

#identify most similar and most different genres
#to find the minimum distance (most similar), we must ignore the diagonal (distance to itself is 0)
np.fill_diagonal(dist_matrix, np.inf)

min_idx = np.unravel_index(np.argmin(dist_matrix), dist_matrix.shape)
max_idx = np.unravel_index(np.argmax(dist_matrix), dist_matrix.shape)

most_similar = (genre_names[min_idx[0]], genre_names[min_idx[1]])
most_diff = (genre_names[max_idx[0]], genre_names[max_idx[1]])
min_dist = dist_matrix[min_idx]
max_dist = dist_matrix[max_idx]

#put the diagonal back to 0 for our function later
np.fill_diagonal(dist_matrix, 0)

print(f"Most similar genres: {most_similar[0].upper()} and {most_similar[1].upper()} (Distance: {min_dist:.2f})")
print(f"Most different genres: {most_diff[0].upper()} and {most_diff[1].upper()} (Distance: {max_dist:.2f})")
print("Musical intuition")
print(f"Does this make sense? Usually, genres like Pop and Latin share very similar upbeat, danceable rhythms, making them mathematically close. Conversely, EDM (heavy, loud, electronic) and R&B or Rock tend to have vastly different acoustic and energy profiles, making them far apart.")

#write recommend_similar_genre function
def recommend_similar_genre(genre_name, top_k=3):
    """Returns the top_k most similar genres based on Euclidean distance."""
    if genre_name not in genre_names:
        return f"Error: Genre '{genre_name}' not found."

    #get the row index for the requested genre
    idx = np.where(genre_names == genre_name)[0][0]

    #get distances for this specific genre
    distances = dist_matrix[idx]

    #sort indices based on distance (index 0 will be the genre itself, so we take 1 to top_k+1)
    sorted_indices = np.argsort(distances)
    top_indices = sorted_indices[1:top_k+1]

    print(f"\nIf you like {genre_name.upper()}, you might also like:")
    for i, sim_idx in enumerate(top_indices, 1):
        print(f"{i}. {genre_names[sim_idx].capitalize()} (Distance: {distances[sim_idx]:.2f})")

recommend_similar_genre('rock', top_k=2)
recommend_similar_genre('edm', top_k=2)
```

### AI Tool Reflection for Class Documentation

* **Prompt Used:** *"Write comprehensive Google-style Python docstrings for a class named `SpotifyAnalyzer` that performs EDA on a Spotify dataset. The class has an `__init__` that loads a URL, and methods for cleaning data, analyzing genre variance, computing NumPy correlations, evaluating high-energy tracks, and generating a dictionary report. Include proper Type Hints (e.g., `Dict[str, Any]`) for all methods."*

```python
import pandas as pd
import numpy as np
import json
from typing import Dict, List, Any

class SpotifyAnalyzer:
    """
    A comprehensive, reusable analysis class for the Spotify Tracks Dataset.

    This class encapsulates the entire Exploratory Data Analysis (EDA) pipeline,
    from data loading and cleaning to advanced NumPy-based correlation and
    statistical analysis.
    """

    def __init__(self, url: str) -> None:
        """
        Initializes the SpotifyAnalyzer, loads the dataset, and defines core columns.

        Args:
            url (str): The raw URL string pointing to the CSV dataset.
        """
        self.df = pd.read_csv(url)
        self.audio_cols: List[str] = [
            'danceability', 'energy', 'loudness', 'speechiness',
            'acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo'
        ]

    def clean_data(self) -> Dict[str, int]:
        """
        Cleans the dataset by removing duplicate tracks based on track_id.

        Returns:
            Dict[str, int]: A dictionary containing the initial and final row counts.
        """
        initial_shape = self.df.shape[0]
        self.df = self.df.drop_duplicates(subset=['track_id'], keep='first')
        return {"initial_rows": initial_shape, "final_rows": self.df.shape[0]}

    def analyze_genre_variance(self) -> Dict[str, Any]:
        """
        Identifies the playlist genre with the highest variance in track popularity.

        Returns:
            Dict[str, Any]: A dictionary containing the genre name and its variance value.
        """
        popularity_var = self.df.groupby('playlist_genre')['track_popularity'].var()
        return {
            "highest_variance_genre": str(popularity_var.idxmax()),
            "variance_value": float(round(popularity_var.max(), 2))
        }

    def analyze_correlations(self) -> Dict[str, Any]:
        """
        Calculates the correlation matrix of audio features using NumPy and extracts
        the highest positive correlation pair.

        Returns:
            Dict[str, Any]: The two most positively correlated features and their score.
        """
        # Vectorized Normalization
        features = self.df[self.audio_cols].to_numpy()
        col_min = np.min(features, axis=0)
        col_max = np.max(features, axis=0)
        norm_features = (features - col_min) / (col_max - col_min)

        # Correlation Matrix
        corr_matrix = np.corrcoef(norm_features, rowvar=False)
        upper_tri_indices = np.triu_indices_from(corr_matrix, k=1)

        max_pos_idx = np.argmax(corr_matrix[upper_tri_indices])
        r_pos = upper_tri_indices[0][max_pos_idx]
        c_pos = upper_tri_indices[1][max_pos_idx]

        return {
            "feature_pair": [self.audio_cols[r_pos], self.audio_cols[c_pos]],
            "correlation_score": float(round(corr_matrix[r_pos, c_pos], 3))
        }

    def evaluate_high_energy(self) -> Dict[str, float]:
        """
        Compares the mean popularity of highly energetic tracks (above µ + 1σ)
        against the overall dataset mean using NumPy boolean masking.

        Returns:
            Dict[str, float]: The mean popularity for both the overall dataset and high-energy tracks.
        """
        energy_array = self.df['energy'].to_numpy()
        pop_array = self.df['track_popularity'].to_numpy()

        threshold = np.mean(energy_array) + np.std(energy_array, ddof=1)
        high_energy_mask = energy_array > threshold

        return {
            "overall_mean_popularity": float(round(np.mean(pop_array), 2)),
            "high_energy_mean_popularity": float(round(np.mean(pop_array[high_energy_mask]), 2))
        }

    def generate_report(self) -> Dict[str, Any]:
        """
        Executes the entire analysis pipeline and aggregates findings into a single report.

        Returns:
            Dict[str, Any]: A nested dictionary of all key insights, ready for JSON serialization.
        """
        report = {
            "Dataset_Cleaning": self.clean_data(),
            "Genre_Analysis": self.analyze_genre_variance(),
            "Audio_Feature_Correlations": self.analyze_correlations(),
            "Energy_vs_Popularity": self.evaluate_high_energy(),
            "Executive_Summary": "High energy tracks outperform average popularity. Loudness and energy are the most tightly correlated features, while pop/latin genres exhibit the most volatile popularity metrics."
        }
        return report

# ==========================================
# EXECUTION BLOCK (Runs instantly with no setup)
# ==========================================
if __name__ == "__main__":
    DATA_URL = "https://raw.githubusercontent.com/rfordatascience/tidytuesday/master/data/2020/2020-01-21/spotify_songs.csv"

    print("Initializing SpotifyAnalyzer...")
    analyzer = SpotifyAnalyzer(DATA_URL)

    print("Generating comprehensive report...\n")
    final_report = analyzer.generate_report()

    # Print the dictionary in a beautiful, readable JSON format
    print(json.dumps(final_report, indent=4))
```
